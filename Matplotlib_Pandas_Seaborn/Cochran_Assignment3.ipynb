```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import csv
```

```python
filename = open('2017.csv', 'r')
file = csv.DictReader(filename)

Economy = []
HappScore = []

for col in file:
    Economy.append(col['Economy..GDP.per.Capita.'])
    HappScore.append(col['Happiness.Score'])

Economy = list(map(float, Economy))
HappScore = list(map(float, HappScore))

plt.scatter(Economy, HappScore, color = 'b', label = 'GDP per Capita')
plt.title("Tyler Cochran - GDP per Capita vs. Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()
```

```python
filename = open('2017.csv', 'r')
file = csv.DictReader(filename)

Freedom = []
HappScore = []

for col in file:
    Freedom.append(col['Freedom'])
    HappScore.append(col['Happiness.Score'])
    
Freedom = list(map(float, Freedom))
HappScore = list(map(float, HappScore))

plt.scatter(Freedom, HappScore, color = 'g', label = 'Freedom')
plt.title("Tyler Cochran - Freedom vs. Happiness Score")
plt.xlabel("Freedom")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()
```

```python
data = pd.read_csv('2017.csv')
data = data.drop(data.columns[0], axis = 1)

x_labels = ['Happiness Rank', 'Happiness Score', 'W_High', 'W_Low', 'Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Trust', 'Dystopia Residual']
y_labels = ['Happiness Rank', 'Happiness Score', 'W_High', 'W_Low', 'Economy', 'Family', 'Health', 'Freedom', 'Generosity', 'Trust', 'Dystopia Residual']
sns.set(font_scale = 0.5)
sns.heatmap(data.corr(), annot = True, fmt = '.1f', xticklabels = x_labels, yticklabels = y_labels, linewidth = .5, linecolor = 'w')
plt.xticks(rotation = 30)
plt.show()
```
